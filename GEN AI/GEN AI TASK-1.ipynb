## Task 1: Replace Hardcoded Prompts

**Goal:** Convert a hardcoded f-string function into a reusable `PromptTemplate`.

```python
# LangChain PromptTemplate approach
# Define the template string using {input_variable} placeholders — no f-strings!
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Function to generate prompt using the template (logic separated from template)
def explain_topic(topic):
    """
    Generates a beginner-friendly explanation prompt for a given topic.
    Uses LangChain PromptTemplate instead of f-strings.
    """
    return explain_template.format(topic=topic)

# Test the function
print("--- Task 1 Output ---")
print(explain_topic("Machine Learning"))
print(explain_topic("Quantum Computing"))
print(explain_topic("Neural Networks"))
```

## Task 2: Multi-Input Prompt System

**Goal:** Build a template with multiple inputs: `topic`, `audience`, and `tone`.

```python
# Define a multi-input PromptTemplate with three variables
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Function to generate multi-input prompts (keeps logic separate from template)
def multi_input_prompt(topic, audience, tone):
    """
    Generates a customized explanation prompt based on topic, audience, and tone.
    """
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)

# Test Cases as specified in the assignment
test_cases = [
    ("AI", "beginners", "friendly"),
    ("Python", "kids", "fun"),
    ("Deep Learning", "engineers", "technical")
]

print("--- Task 2 Output ---")
for topic, audience, tone in test_cases:
    print(multi_input_prompt(topic, audience, tone))
```

## Task 3: Prompt Variations Engine

**Goal:** Create 3 different `PromptTemplates` - Teaching, Interview, and Storytelling - and run them for a single topic.

```python
# Define 3 distinct PromptTemplates for different prompt styles

# Template 1: Teaching style — step-by-step explanation
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Template 2: Interview style — generates interview questions
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Template 3: Storytelling style — explains through a narrative
storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Store templates in a dictionary for easy access
variation_templates = {
    "teaching": teaching_template,
    "interview": interview_template,
    "storytelling": storytelling_template
}

# Function to generate all variations for a given topic
def prompt_variations_engine(topic):
    """
    Generates prompts in all 3 styles (teaching, interview, storytelling)
    for a given topic.
    """
    results = {}
    for style, template in variation_templates.items():
        results[style] = template.format(topic=topic)
    return results

# Test with topic = Machine Learning
print("--- Task 3 Output ---")
variations = prompt_variations_engine("Machine Learning")
for style, prompt in variations.items():
    print(f"[{style.upper()}]: {prompt}")
```

## Task 4: ChatPromptTemplate System

**Goal:** Build a chat-based prompt system using `ChatPromptTemplate` with System and User roles.

```python
# Define system behavior descriptions for each role
# Each role shapes how the AI model will respond to the user's topic
role_descriptions = {
    "teacher": "You are a knowledgeable and patient teacher. Explain concepts clearly with examples.",
    "interviewer": "You are a professional interviewer. Ask insightful questions and evaluate answers.",
    "motivator": "You are an enthusiastic motivator. Inspire and encourage learners with positivity."
}

def build_chat_prompt(role, topic):
    """
    Builds a ChatPromptTemplate dynamically based on role and topic.
    - System message: defines the assistant's behavior/persona
    - Human message: the actual user query
    """
    # Validate that the role exists
    if role not in role_descriptions:
        raise ValueError(f"Invalid role '{role}'. Choose from: {list(role_descriptions.keys())}")

    # System message sets the AI's persona (uses the role description)
    system_message = SystemMessagePromptTemplate.from_template(
        role_descriptions[role]
    )

    # Human message contains the dynamic topic input
    human_message = HumanMessagePromptTemplate.from_template(
        "Please help me understand {topic}."
    )

    # Combine system + human into a full ChatPromptTemplate
    chat_template = ChatPromptTemplate.from_messages([system_message, human_message])

    # Format the template with the actual topic value
    formatted_messages = chat_template.format_messages(topic=topic)
    return formatted_messages

# Test with role = 'teacher' and topic = 'Neural Networks'
print("--- Task 4 Output ---")

test_scenarios = [
    ("teacher", "Neural Networks"),
    ("interviewer", "Machine Learning"),
    ("motivator", "Deep Learning")
]

for role, topic in test_scenarios:
    print(f"\nRole: {role.upper()} | Topic: {topic}")
    messages = build_chat_prompt(role, topic)
    for msg in messages:
        print(f"  [{msg.type.upper()}]: {msg.content}")
```

## Task 5: Input Validation Layer

**Goal:** Create a `validate_inputs` function that enforces allowed values for `audience` and `tone`.

```python
# Define the allowed values for each input parameter
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES = ["formal", "casual", "fun"]

# Default fallback values (used when an invalid input is provided)
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE = "casual"

def validate_inputs(audience, tone, strict=False):
    """
    Validates audience and tone inputs against allowed values.

    Parameters:
        audience (str): Target audience level.
        tone (str): Communication tone.
        strict (bool): If True, raises ValueError on invalid input.
                       If False, assigns default values (default behavior).

    Returns:
        tuple: (validated_audience, validated_tone)
    """
    # Normalize inputs to lowercase to allow case-insensitive matching
    audience = audience.lower().strip()
    tone = tone.lower().strip()

    # Validate audience
    if audience not in VALID_AUDIENCES:
        if strict:
            raise ValueError(f"Invalid audience '{audience}'. Must be one of {VALID_AUDIENCES}")
        else:
            print(f" Invalid audience '{audience}'. Defaulting to '{DEFAULT_AUDIENCE}'.")
            audience = DEFAULT_AUDIENCE

    # Validate tone
    if tone not in VALID_TONES:
        if strict:
            raise ValueError(f"Invalid tone '{tone}'. Must be one of {VALID_TONES}")
        else:
            print(f" Invalid tone '{tone}'. Defaulting to '{DEFAULT_TONE}'.")
            tone = DEFAULT_TONE

    return audience, tone

# Test Cases
print("--- Task 5 Output ---")

# Test 1: Valid inputs
print("\nTest 1 — Valid inputs:")
result = validate_inputs("expert", "formal")
print(f"  Validated: {result}")

# Test 2: Invalid inputs — defaults assigned
print("\nTest 2 — Invalid inputs (default mode):")
result = validate_inputs("student", "angry")
print(f"  Validated: {result}")

# Test 3: Invalid inputs — strict mode raises error
print("\nTest 3 — Invalid inputs (strict mode):")
try:
    result = validate_inputs("student", "angry", strict=True)
except ValueError as e:
    print(f"  ValueError caught: {e}")
```

## Task 6: Prompt Generator App

**Goal:** Create a `generate_prompt` function that uses validation + PromptTemplate to generate style-specific prompts.

```python
# Define style-specific PromptTemplates for the generator app
# Each style has a distinct template structure
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Generate interview questions about {topic} for {audience} in a {tone} tone"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}

def generate_prompt(topic, audience, tone, style):
    """
    Generates a structured, validated prompt using LangChain PromptTemplate.

    Parameters:
        topic (str): Subject to explain.
        audience (str): Target audience (beginner / intermediate / expert).
        tone (str): Communication tone (formal / casual / fun).
        style (str): Prompt style (teaching / interview / storytelling).

    Returns:
        str: The fully formatted prompt string.
    """
    # Step 1: Validate audience and tone using our validation layer (Task 5)
    validated_audience, validated_tone = validate_inputs(audience, tone)

    # Step 2: Validate style
    style = style.lower().strip()
    if style not in style_templates:
        print(f" Invalid style '{style}'. Defaulting to 'teaching'.")
        style = "teaching"

    # Step 3: Select the appropriate PromptTemplate based on style
    selected_template = style_templates[style]

    # Step 4: Format and return the final prompt
    return selected_template.format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )

# Test Cases
print("--- Task 6 Output ---")

# Example from assignment brief
print("\nExample 1:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))

print("\nExample 2:")
print(generate_prompt("Python", "intermediate", "formal", "teaching"))

print("\nExample 3 (invalid inputs):")
print(generate_prompt("AI Ethics", "student", "cool", "interview"))
```

## Task 7: Template Reusability Test (10%)

**Goal:** Demonstrate that ONE template can produce 5 different outputs (same structure, different content).

```python
# One single reusable PromptTemplate — the core principle of this assignment
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to a {audience} using a {tone} tone"
)

# 5 different input sets to demonstrate reusability
# Same template → different outputs each time
test_inputs = [
    {"topic": "Artificial Intelligence",  "audience": "beginner",     "tone": "friendly"},
    {"topic": "Data Structures",          "audience": "student",      "tone": "casual"},
    {"topic": "Blockchain",               "audience": "executive",    "tone": "formal"},
    {"topic": "Reinforcement Learning",   "audience": "researcher",   "tone": "technical"},
    {"topic": "Cloud Computing",          "audience": "entrepreneur", "tone": "inspiring"},
]

print("--- Task 7 Output — Template Reusability Test ---")
print("Template used:", reusable_template.template)
print("-" * 55)

# Apply the single template to all 5 inputs
for i, inputs in enumerate(test_inputs, start=1):
    output = reusable_template.format(**inputs)
    print(f"Input {i}: {output}")
```

## Full Pipeline Demo

**Flow:** User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output

```python
def run_pipeline(topic, audience, tone, style):
    """
    Full end-to-end prompt pipeline:
    User Input → Validation → Template Selection → Prompt Generation → Output
    """
    print("=" * 55)
    print(" USER INPUT")
    print(f"   topic={topic}, audience={audience}, tone={tone}, style={style}")

    print("\n VALIDATION")
    validated_audience, validated_tone = validate_inputs(audience, tone)
    print(f"   audience={validated_audience}, tone={validated_tone}")

    print("\n TEMPLATE SELECTION & FORMATTING")
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)

    print("\n GENERATED PROMPT")
    print(f"   {final_prompt}")
    print("=" * 55)
    return final_prompt

# Run the full pipeline
print("--- Full Pipeline Demo ---\n")
run_pipeline("Large Language Models", "beginner", "fun", "storytelling")
run_pipeline("Python Decorators", "advanced", "technical", "teaching")
```
