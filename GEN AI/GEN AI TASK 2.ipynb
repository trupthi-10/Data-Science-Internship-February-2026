# Task-3 AI Resume Screening System with Tracing


```python
def extract_skills(resume):
    """
    Extract skills, tools, and experience from resume text
    (simple keyword-based logic)
    """

    skills_list = ["Python", "Machine Learning", "Deep Learning", "SQL", "Data Visualization"]
    tools_list = ["Pandas", "NumPy", "Matplotlib", "TensorFlow", "Scikit-learn"]

    found_skills = [skill for skill in skills_list if skill.lower() in resume.lower()]
    found_tools = [tool for tool in tools_list if tool.lower() in resume.lower()]

    # Extract experience (simple logic)
    experience = 0
    words = resume.split()
    for i, word in enumerate(words):
        if word.isdigit() and i + 1 < len(words) and "year" in words[i + 1].lower():
            experience = int(word)

    return {
        "Skills": found_skills,
        "Tools": found_tools,
        "Experience": experience
    }
```

```python
def match_skills(resume_data, job_description):
    """
    Compare resume with job description
    """

    jd_lower = job_description.lower()

    matching = []
    missing = []

    for skill in resume_data["Skills"]:
        if skill.lower() in jd_lower:
            matching.append(skill)

    # Check missing skills from JD
    jd_skills = ["Python", "Machine Learning", "Deep Learning", "SQL", "Data Visualization"]

    for skill in jd_skills:
        if skill not in matching:
            missing.append(skill)

    return {
        "Matching Skills": matching,
        "Missing Skills": missing
    }
```

```python
def calculate_score(match_data, experience):
    """
    Assign score based on matching and experience
    """

    match_count = len(match_data["Matching Skills"])

    # Base score from skills
    score = match_count * 15

    # Add experience bonus
    score += experience * 5

    # Cap score at 100
    score = min(score, 100)

    return score
```

```python
def generate_explanation(match_data, score):
    """
    Generate explanation for score
    """

    explanation = f"""
Candidate Evaluation:

Score: {score}

Strengths:
- Matching Skills: {', '.join(match_data["Matching Skills"])}

Weaknesses:
- Missing Skills: {', '.join(match_data["Missing Skills"])}

Final Reason:
The score is based on the number of matching skills and experience level.
"""

    return explanation
```

```python
def evaluate_resume(resume, job_description):
    """
    Complete pipeline:
    Extract → Match → Score → Explain
    """

    # Step 1: Extract
    extracted = extract_skills(resume)

    # Step 2: Match
    matched = match_skills(extracted, job_description)

    # Step 3: Score
    score = calculate_score(matched, extracted["Experience"])

    # Step 4: Explain
    explanation = generate_explanation(matched, score)

    return {
        "Extracted": extracted,
        "Matched": matched,
        "Score": score,
        "Explanation": explanation
    }
```

```python
job_description = """
Looking for a Data Scientist with:
Python, Machine Learning, Deep Learning, SQL, Data Visualization
Minimum 2 years experience
"""
```

```python
strong_resume = """
Python developer with 3 years experience.
Skilled in Machine Learning, Deep Learning, SQL, Pandas, Matplotlib.
"""

average_resume = """
Software engineer with 2 years experience.
Knows Python, SQL and basic Machine Learning.
"""

weak_resume = """
Fresher with basic programming knowledge.
Knows C language only.
"""
```

```python
print("===== STRONG =====")
print(evaluate_resume(strong_resume, job_description))

print("\n===== AVERAGE =====")
print(evaluate_resume(average_resume, job_description))

print("\n===== WEAK =====")
print(evaluate_resume(weak_resume, job_description))
```
