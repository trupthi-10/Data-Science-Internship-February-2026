**1. Mobile Recharge Validation System**

```python
def validate_plan(amount, valid_plans):
    if amount in valid_plans:
        return True
    else:
        return False


valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: "))

        # Step 1: Check minimum amount
        if amount < 50:
            print("Recharge amount must be at least ₹50\n")
            continue

        # Step 2: Show plans only if amount >= 50
        print("Available Plans:", valid_plans)

        selected_plan = int(input("Select a plan from above: "))

        if validate_plan(selected_plan, valid_plans):
            print("Recharge Successful")
            break
        else:
            print("Invalid plan selected. Please try again.\n")

    except ValueError:
        print("Please enter valid numeric input.\n")
```

**2. Inventory Reorder Alert System**

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert ⚠")
        else:
            print(f"{item}: Stock OK")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 30
}

check_inventory(inventory)
```

 **3. Student Result Processing System**

```python
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 55, 40, 70]

result = process_result(marks_list)
print("Result:", result)
```

**4. Cab Fare Estimator**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(km, peak_hour)
        print("Total Fare: ₹", round(total_fare, 2))

        break

    except ValueError:
        print("Invalid input. Please try again.\n")
```

**5. Attendance Eligibility Checker**



```python
def check_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P"]

status = check_eligibility(attendance_list)
print("Status:", status)
```

**6. Password Strength Checker**

```python
def check_password(password):
    if len(password) < 8:
        return "Password too short it should be more that 8 charecters"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if not has_digit:
        return "Password must contain at least one digit"

    if not has_special:
        return "Password must contain at least one special character (@#$)"

    return "Strong"


# Retry loop
while True:
    user_password = input("Enter password: ")
    result = check_password(user_password)

    if result == "Strong":
        print("Strong Password")
        break
    else:
        print(result)
        print("Please try again.\n")
```
