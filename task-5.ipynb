**1. Social Media – Post Engagement Analyzer**

```python
def analyze_post(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes_list = []

n = int(input("Enter number of like entries: "))

for i in range(n):
    value = int(input(f"Enter likes {i+1}: "))
    likes_list.append(value)

analyze_post(likes_list)
```

 **3. Agriculture – Rainfall Adequacy Checker**

```python
def check_rainfall(rainfall_data, required_level):

    total = 0

    for value in rainfall_data:
        total += value

    average = total / len(rainfall_data)

    print("Average Rainfall:", round(average, 2))

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rainfall_list = []

n = int(input("Enter number of seasonal rainfall readings: "))

for i in range(n):
    value = float(input(f"Enter rainfall reading {i+1}: "))
    rainfall_list.append(value)

required = float(input("Enter required rainfall level: "))

check_rainfall(rainfall_list, required)
```

**4.Social Media – Duplicate Account Detection**

```python
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# User Input
user_list = []
n = int(input("Enter number of usernames: "))

for i in range(n):
    name = input(f"Enter username {i+1}: ")
    user_list.append(name)

detect_duplicates(user_list)
```

**6. Agriculture – Premium Crop Price Filter**

```python
def filter_crop_prices(prices):
    premium = []
    regular = []

    for price in prices:
        if price > 2000:
            premium.append(price)
        else:
            regular.append(price)

    return premium, regular


# Taking user input
n = int(input("Enter number of crop prices: "))

price_list = []
for i in range(n):
    price = float(input(f"Enter price {i+1}: "))
    price_list.append(price)

premium_crops, regular_crops = filter_crop_prices(price_list)

print("Premium Crops:", premium_crops)
print("Regular Crops:", regular_crops)
```

**7. System Monitoring – Application Health Checker**

**9. E-Learning – Student Attendance Eligibility System**

```python
def check_attendance(attendance_list):
    total = 0

    for value in attendance_list:
        total += value

    percentage = total / len(attendance_list)

    print("Attendance Percentage:", round(percentage, 2))

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# User Input
attendance = []
n = int(input("Enter number of days attendance recorded: "))

for i in range(n):
    value = float(input(f"Enter attendance percentage for day {i+1}: "))
    attendance.append(value)

check_attendance(attendance)
```

**10. Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):
    bill = 0

    # Slab Calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# User Input
units = int(input("Enter electricity units consumed: "))

total_bill, usage_status = calculate_bill(units)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
