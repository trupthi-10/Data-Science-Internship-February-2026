**Problem 1: Smart Parking Lot Management System**

```python
def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for entry in logs:
        if entry == "IN":
            current += 1
        elif entry == "OUT" and current > 0:
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


capacity = int(input("Enter Parking Capacity: "))
logs = input("Enter Vehicle Logs (comma separated IN/OUT): ").split(",")

logs = [log.strip().upper() for log in logs]

manage_parking(capacity, logs)
```

**Problem 2: Online Food Delivery Time Estimator**

```python
def estimate_delivery(distance, traffic, weather):
    eta = distance * 5   # Base 5 minutes per km

    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 8

    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# User Input
distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low/Medium/High): ")
weather = input("Enter Weather (Clear/Rainy/Storm): ")

estimate_delivery(distance, traffic, weather)
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**


```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy, 2), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# User Input
total_seats = int(input("Enter Total Seats: "))
booked_count = int(input("Enter Number of Booked Seats: "))

booked_seats = [1] * booked_count

analyze_occupancy(total_seats, booked_seats)
```

**Problem 4: Cloud Server Load Classification System**


```python
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average, 2), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# User Input
readings = input("Enter CPU readings separated by commas: ").split(",")
cpu_readings = [float(r.strip()) for r in readings]

classify_server_load(cpu_readings)
```

**Problem 5: Smart Classroom Resource Usage Monitor**


```python
def monitor_resources(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# User Input
resources = {}

n = int(input("Enter number of resources: "))

for i in range(n):
    name = input("Enter Resource Name: ")
    hours = int(input("Enter Usage Hours: "))
    resources[name] = hours

monitor_resources(resources)
```

**Problem 6: Online Event Registration Capacity Controller**


```python
def manage_registrations(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# User Input
capacity = int(input("Enter Event Capacity: "))
registrations = int(input("Enter Total Registrations: "))

manage_registrations(capacity, registrations)
```
