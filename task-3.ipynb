**Problem Statement 1: Employee Performance Bonus Eligibility**

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find all employees who have highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

**Problem Statement 2: Search Query Keyword Analysis**


```python
query = "Buy mobile phone buy phone online"

query = query.lower()

# Remove punctuation
for char in [".", ",", "!", "?"]:
    query = query.replace(char, "")

words = query.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Display only words appearing more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

**Problem Statement 3: Sensor Data Validation**


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Even readings are valid
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**Problem Statement 4: Email Domain Usage Analysis**


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate and display percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

**Problem Statement 5: Sales Spike Detection**


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.30  # 30% above average

# Detect spike days
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

**Problem Statement 6: Duplicate User ID Detection**


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

# Count occurrences
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Display duplicates only
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
